```python
import numpy as np
```

step 1:
Encoding and separating the data using numpy 

```python
data = [
    [150, 7.0, 1, 'Apple'],
    [120, 6.5, 0, 'Banana'],
    [180, 7.5, 2, 'Orange'],
    [155, 7.2, 1, 'Apple'],
    [110, 6.0, 0, 'Banana'],
    [190, 7.8, 2, 'Orange'],
    [145, 7.1, 1, 'Apple'],
    [115, 6.3, 0, 'Banana']
]
data
```

```python
labels = {"Apple" : 0, "Banana" : 1,"Orange" : 2}
```

```python
X = np.array([x[0:3] for x in data])
print(X)


#normalising X:
mean= np.mean(X , axis = 0)                                             #axis = 0 means mean col wise ie mean per feature while axis = 1 means mean row-wise
std= np.std(X , axis = 0)
X_scaled = (X - mean)/std
print(X_scaled)
```

```python
y = np.array([labels[x[-1]] for x in data])
y
```

STEP 2: WRITING THE EUCLIDEAN FUNCTION

```python
def euclidean_distance (a,b):
    return np.sqrt(np.sum((a-b)**2))
    
```

STEP 3 : MAKING THE CLASS FOR KNN

```python
class KNN: 
    def __init__(self,k = 3):
        self.k = k

    def fit (self,X,y):
        self.X = X
        self.y = y
    
    def predict_one(self,x):
        distances = [euclidean_distance(y,x) for y in self.X]
        indices = np.argsort(distances)
        nearest_k = indices[0:self.k]
        nearest_k_labels = self.y[nearest_k]
        values , counts = np.unique(nearest_k_labels,return_counts=True)
        max_index = np.argmax(counts)
        max_value = values[max_index]
        return max_value
    
    def predict (self,X_test):
        return [self.predict_one(t) for t in X_test]

```

STEP 4:TESTING THE SAMPLE

```python
test_data = np.array([
    [118, 6.2 , 0],  # Expected: Banana
    [160, 7.3, 1],  # Expected: Apple
    [185, 7.7, 2]   # Expected: Orange
])
model = KNN(k=3)
model.fit (X_scaled,y)

#normalising the test_data :

test_data_scaled = (test_data - mean)/std



reverse_labels = {v: k for k, v in labels.items()}
ans = [reverse_labels[x] for x in model.predict(test_data_scaled)]
print(ans)
```
